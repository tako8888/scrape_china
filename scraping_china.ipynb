# Now, make it a function to read html files
## def read_html return an array

```python
def read_html(file):
    soup = BeautifulSoup(file)
    projects_section = soup.select(".feeds-container")[0]
    project_els = projects_section.select(".note-item")

    answer_list = []

    for i, elem in enumerate(project_els):
        try:
            clean_link = elem.select('a[style="display: none;"]')[0]['href']
            dirty_link = elem.select('a[target="_self"]')[0]['href']
            try:
                title = elem.select('.footer>.title')[0].text
            except IndexError:
                title = ""
            like_count = elem.select('.count')[0].text
            author_name = elem.select('.name')[0].text
            author_link = elem.select('.author')[0]['href']
            temp_dictionary = {"title":title, "like_count":like_count, "author_name":author_name, "author_link":author_link, "clean_link":clean_link, "dirty_link":dirty_link}
            answer_list.append(temp_dictionary)
            
        except IndexError:
            print(i, elem.select('.query-note-header-text'))
    
    return answer_list


```

```python
directory = "./data/scraped/"

temp_array = []

for filename in os.listdir(directory):
    if filename.endswith('.html'):
        full_path = os.path.join(directory, filename)
        with open(full_path, 'r') as file: 
            content = file.read()
        temp_array.extend(read_html(content))

```

```python
import re
# 1 万 = 10000
def correct_likes(value):
    match = re.search(r'\d+', value)
    if match:
        number = int(match.group())

        if '万' in value:
            answer = number * 10000
        
        elif value == '赞':
            answer = 0
        else:
            answer = number
    else:
        answer = 0
        print('there are no numbers')
    return answer

df['normalized_count'] = df['like_count'].apply(correct_likes)
```

```python
from datetime import timedelta

# 基準日を手動で設定
base_date = datetime(2024, 7, 21)


def normalize_date(value):
    if pd.isna(value):
        return None

    # '编辑于' というテキストを取り除き、その後の地名を取り除く
    cleaned_value = re.sub(r'编辑于\s*', '', value)
    # 日付表現後の余分なテキスト（地名など）を除去
    cleaned_value = re.sub(r'(\d{2}-\d{2}|\d{4}-\d{2}-\d{2}|\d+ 天前|昨天|今天)(\s+\S+)?$', r'\1', cleaned_value).strip()

    # YYYY-MM-DD 形式
    if re.match(r'\d{4}-\d{2}-\d{2}', cleaned_value):
        return cleaned_value[:10]
    
    # MM-DD 形式
    elif re.match(r'\d{2}-\d{2}', cleaned_value):
        return f'2024-{cleaned_value}'  # 2024年のデータと仮定して、年を追加
    
    # 相対日付の解析
    elif '天前' in cleaned_value:
        days_ago = int(re.search(r'\d+', cleaned_value).group())
        new_date = base_date - timedelta(days=days_ago)
        return new_date.strftime('%Y-%m-%d')
    
    elif '昨天' in cleaned_value:
        new_date = base_date - timedelta(days=1)
        return new_date.strftime('%Y-%m-%d')
    
    elif '今天' in cleaned_value:
        return base_date.strftime('%Y-%m-%d')

    return None

merged_df['part_date'] = merged_df['date'].apply(normalize_date)
merged_df[merged_df['part_date'].isna()]
```
